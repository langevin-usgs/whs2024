```python
shb.shape, sha.shape
```

```python
diff = shb - sha
ai = plt.imshow(diff)
plt.colorbar(ai)
```

```python
diff.min(), diff.max()
```

```python
np.where(diff > 0)
```

```python
idx = np.where(diff > 0)
diff_positive = np.zeros(diff.shape)
diff_positive[idx] = diff[idx]
plt.imshow(diff_positive)
```

```python
idx = np.where(diff < 0)
diff_negative = np.zeros(diff.shape)
diff_negative[idx] = diff[idx]
plt.imshow(diff_negative)
```

```python
dp = diff_positive.sum() * 500 ** 2
```

```python
dn = diff_negative.sum() * 500 ** 2
```

```python
nrow, ncol = shb.shape
nrow, ncol
x = np.arange(ncol)
y = shb[int(nrow / 2), :]
plt.plot(x, y)
```

```python
nrow, ncol = shb.shape
nrow, ncol
x = np.arange(ncol)
row_to_plot = int(nrow / 2) - 50
y1 = shb[row_to_plot, :]
y2 = sha[row_to_plot, :]
plt.plot(x, y1, label="Before")
plt.plot(x, y2, label="After")
plt.legend()
```

```python
# ax is upper plot
ax = plt.subplot(2, 1, 1)
ax.plot(x, y1, "ro--")
ax.set_title("Before")
ax.set_ylim(1200, 3000)

# ax is lower plot
ax = plt.subplot(2, 1, 2)
ax.plot(x, y2)
ax.set_title("After")
ax.set_ylim(1200, 3000)
plt.tight_layout()
```

```python
nrow, ncol = shb.shape
nrow, ncol
x = np.arange(ncol)
row_to_plot = int(nrow / 2) - 50
y1 = shb[row_to_plot, :]
y2 = sha[row_to_plot, :]
plt.plot(x, y1, label="Before")
plt.plot(x, y2, label="After")

ax = plt.gca()
ax.set_ylim(1200, 3000)
```

```python
nrow, ncol = shb.shape
nrow, ncol
x = np.arange(ncol)
row_to_plot = int(nrow / 2) - 50
y1 = shb[row_to_plot, :]
y2 = sha[row_to_plot, :]
plt.fill_between(x, y1, y2)

ax = plt.gca()
ax.set_ylim(1200, 3000)
ax.set_xlabel("Distance in feet")
ax.set_ylabel("Elevation in feet")
ax.text(150, 2800, "the peak")
```

```python
nrow, ncol = shb.shape
nrow, ncol
ax = plt.subplot(1, 1, 1)
x = np.arange(ncol)
row_to_plot = int(nrow / 2) - 50
y1 = shb[row_to_plot, :]
y2 = sha[row_to_plot, :]
ax.fill_between(x, y1, y2)

ax.set_ylim(1200, 3000)
ax.set_xlabel("Distance in feet")
ax.set_ylabel("Elevation in feet")
ax.text(150, 2800, "the peak")
plt.savefig('myfig.png', dpi=300)
```
